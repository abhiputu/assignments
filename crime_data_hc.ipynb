```python
import pandas as pd
import numpy as np
import scipy as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/C5295630/Documents/DataScience/Assignment/crime_data.csv")
data.head()
```

```python
data = data.rename({'Unnamed: 0':'city_name'},axis=1)
```

```python
## Normalised the data
def norm_fun(i):
    x = (i-i.min())-(i.max()-i.min())
    return(x)
```

```python
df_data = norm_fun(data.iloc[:,1:])
df_data.head()
```

```python
dendrogram = sch.dendrogram(sch.linkage(df_data,method ='single'))
```

```python
hc = AgglomerativeClustering(n_clusters = 3,affinity = 'euclidean',linkage='single')
```

```python
y_hc = hc.fit_predict(df_data)
y_hc
```

```python
Clusters = pd.DataFrame(y_hc,columns=['Clusters'])
Clusters.head(10)
```

```python
data['h_clusterid'] = hc.labels_
```
